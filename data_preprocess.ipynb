필요한 컬럼만 남기기

```python
import pandas as pd

df = pd.read_csv(r".\src\assets\netflix_korea_popular_35.csv")

# 필요한 컬럼만 선택
df = df[["title", "type", "listed_in", "imdb_rating", "description", "release_year"]]

print(df.head())
print(df.info())
```

결측치 처리 및 imdb_rating 숫자형인지 확인

```python
df = df.dropna(subset=["type", "listed_in", "imdb_rating"])

df["imdb_rating"] = pd.to_numeric(df["imdb_rating"], errors="coerce")
df = df.dropna(subset=["imdb_rating"])
```

장르 문자열을 다루기 좋게 공백 정리

```python
df["listed_in"] = df["listed_in"].str.strip()
```

전처리 파일 저장

```python
df.to_csv(r".\src\assets\netflix_korea_popular_35_preprocessed.csv", index=False, encoding="utf-8-sig")
```
